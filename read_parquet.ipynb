```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("ReadParquet")
    .master("local[*]")
    .getOrCreate()
)
```

```python
spark
```

```python

df1 = spark.read \
    .format("parquet") \
    .load("fund_metadata_trail")

df1.show()
```

```python
from pyspark.sql.functions import input_file_name
df1.withColumn("Path" , input_file_name()).show()
```

```python
df1.createOrReplaceTempView("fund")
```

```python
sql  = """
Select ABC.* from (
    select t.*,
    row_number() over(partition by t.fund_id order by t.effective_date desc) RN
    from fund t
) ABC where ABC.RN = 1
"""

spark.sql(sql).show()
```
